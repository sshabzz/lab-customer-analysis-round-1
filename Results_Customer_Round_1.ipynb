# Correct the values in the column 'customer lifetime value'.

(only 2 of the files need this step.)

```python
#file1['customer_lifetime_value'] = pd.to_numeric(file1['customer_lifetime_value'].apply(lambda x: float(x)/100))
#file1['customer_lifetime_value'] = file1['customer_lifetime_value'] / 100
file1['customer_lifetime_value'] = pd.to_numeric(file1['customer_lifetime_value'].astype(str).str.replace('%',''), errors='coerce')/100
file1
```
